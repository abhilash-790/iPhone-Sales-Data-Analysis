Top 10 highest rated iphone on flipkart in india?

```python
df.head()
```

```python
 highest_rated = df.sort_values(by=['Star Rating'],ascending=False)
 highest_rated = highest_rated.head(10)
 print(highest_rated['Product Name'])
```

```python
print(highest_rated)
```

**How many ratings do the highest- rated iphone on Flipcart have?**

```python
 iphone = highest_rated['Product Name'].value_counts()
 label = iphone.index
 counts = highest_rated['Number Of Ratings']
 figure = px.bar(highest_rated, x=label, y=counts)
 figure.show()
```

**Which iphone has the highest number of reviews on Flipcart?**

```python
df.head()
```

```python
 iphone = highest_rated['Product Name'].value_counts()
 label = iphone.index
 counts = highest_rated['Number Of Reviews']
 figure = px.bar(highest_rated, x=label, y=counts, title='Number of Reviews of highest rated iphone')
 figure.show()
```

**What is the relationship between the discount percentage and the number of ratings on Flipcart?**

```python
figure = px.scatter(data_frame = df, x='Discount Percentage',
                    y='Discount Percentage', size='Sale Price',
                    trendline='ols', title='Relationship between Discount Percentage and Sale Price')
figure.show()
```

**Most expensive and least expensive**
